```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA

# Load and inspect
df = pd.read_csv('SpotifyFeatures.csv')
print(df.shape)
print(df.isnull().sum())

# Choose features
features = ['danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
X = df[features]

# Handle missing values
X = X.dropna()  # You could also use fillna() if needed

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow method to find optimal K
inertia = []
k_range = range(2, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()

# KMeans clustering with chosen k
kmeans = KMeans(n_clusters=4, random_state=42)
k_labels = kmeans.fit_predict(X_scaled)
df.loc[X.index, 'KMeans_Cluster'] = k_labels

# DBSCAN clustering
dbscan = DBSCAN(eps=1.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)
df.loc[X.index, 'DBSCAN_Cluster'] = db_labels

# Evaluation
print("K-Means Silhouette Score:", silhouette_score(X_scaled, k_labels))
print("K-Means DB Index:", davies_bouldin_score(X_scaled, k_labels))

# Filter DBSCAN noise for evaluation
mask = db_labels != -1
if np.any(mask):
    print("DBSCAN Silhouette Score:", silhouette_score(X_scaled[mask], db_labels[mask]))
    print("DBSCAN DB Index:", davies_bouldin_score(X_scaled[mask], db_labels[mask]))
else:
    print("DBSCAN produced only noise (no core clusters to evaluate).")

# Visualize with PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 4))

# KMeans visualization
plt.subplot(1, 2, 1)
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=k_labels, palette='Set2')
plt.title('K-Means Clusters')

# DBSCAN visualization
plt.subplot(1, 2, 2)
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=db_labels, palette='Set2')
plt.title('DBSCAN Clusters')

plt.tight_layout()
plt.show()
```
